```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    i = 3
    while i * i <= num:
        if num % i == 0:
            return False
        i += 2
    return True

print(f"Is 13 prime? {is_prime(13)}")
print(f"Is 4 prime? {is_prime(4)}")
print(f"Is 2 prime? {is_prime(2)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 7 prime? {is_prime(7)}")
```

```python
class Dog:    #blueprint or template to create objects(dogs)
  def init_(self,name,age):     #self refers to the current object
    self.name=name      #attribute(variable) called name
    self.age=age   #2nd object
```

```python
class Dog:
  def __init__(self,name,age):
    self.name=name
    self.age=age
ozzy=Dog("Ozzy",2)
print(ozzy.name)
print(ozzy.age)
print(ozzy.name+"is"+str(ozzy.age)+"years old.")
```
