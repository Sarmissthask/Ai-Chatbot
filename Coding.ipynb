```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    i = 3
    while i * i <= num:
        if num % i == 0:
            return False
        i += 2
    return True

print(f"Is 13 prime? {is_prime(13)}")
print(f"Is 4 prime? {is_prime(4)}")
print(f"Is 2 prime? {is_prime(2)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 7 prime? {is_prime(7)}")
```

```python
import pandas as pd
import yfinance as yf

# Download Apple stock data
aapl = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

short_window = 40
long_window = 100

signals = pd.DataFrame(index=aapl.index)
signals['signal'] = 0.0

signals['short_mavg'] = aapl['Close'].rolling(window=short_window, min_periods=1).mean()
signals['long_mavg'] = aapl['Close'].rolling(window=long_window, min_periods=1).mean()

signals['positions'] = signals['signal'].diff()

print(signals)
```
